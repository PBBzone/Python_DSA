Question 3 of HackerRank 

```python
import math
import os
import random
import re
import sys

#
# Complete the 'compareTriplets' function below.
#
# The function is expected to return an INTEGER_ARRAY.
# The function accepts following parameters:
#  1. INTEGER_ARRAY a
#  2. INTEGER_ARRAY b
#

def compareTriplets(a, b):
    # Initialize scores
    alice = 0
    bob = 0
    
    # Compare each category
    for i in range(3):
        if a[i] > b[i]:
            alice += 1
        elif a[i] < b[i]:
            bob += 1
            
    # Return the results as an array/list
    return [alice, bob]

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    a = list(map(int, input().rstrip().split()))

    b = list(map(int, input().rstrip().split()))

    result = compareTriplets(a, b)

    fptr.write(' '.join(map(str, result)))
    fptr.write('\n')

    fptr.close()
```

Question 5 of HackerRan

```python
#!/bin/python3

import math
import os
import random
import re
import sys

#
# Complete the 'timeConversion' function below.
#
# The function is expected to return a STRING.
# The function accepts STRING s as parameter.
#

def timeConversion(s):
    # Extract the AM/PM modifier and the time components
    period = s[-2:]
    hour = s[:2]
    rest_of_time = s[2:-2]
    
    if period == "AM":
        # If it's AM, 12 becomes 00. Everything else stays the same.
        if hour == "12":
            hour = "00"
    else:
        # If it's PM, add 12 to the hour, unless it's already 12.
        if hour != "12":
            hour = str(int(hour) + 12)
            
    # Recombine the formatted hour with the rest of the time string
    return hour + rest_of_time

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    s = input()

    result = timeConversion(s)

    fptr.write(result + '\n')

    fptr.close()
```
